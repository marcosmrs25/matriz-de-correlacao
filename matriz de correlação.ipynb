```python
#impotando bibliotecas 
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#carregando dados
jura = pd.read_csv('jura.pred.csv', sep = ';')

#visualizar dados
jura
 
```

```python
#criando a correlação
corr = jura[['Cd','Co', 'Cr','Cu', 'Ni', 'Pb', 'Zn']]
```

```python
#criar matriz de correlação
mask = np.zeros_like(corr, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize = (11,9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap =cmap, vmax=1, vmin=0, center=0, square=True, linewidth=.5,cbar_kws={"shrink":.5})
corr = jura.corr()
```
